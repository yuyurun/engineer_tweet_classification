```python
def tweet_save_user_keyinprofile(tweets,filepath,keyword):
    user = []
    with open(filepath + "/tweet_user_"  + keyword + ".txt","a") as f:
        for tweet in tweets:
            sentence = tweet.author
            if re.search(keyword,sentence.description):
                print(sentence.description)
                f.write(str(sentence.screen_name))
                f.write("\n")
                user.append(str(sentence.screen_name))
    return user
```
